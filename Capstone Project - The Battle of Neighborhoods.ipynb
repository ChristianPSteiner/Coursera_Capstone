#### Lets explore the content of reason items, perhaps we find feedback about the different venue's popularity

```python
# filter reason items to learn if content can be used as filter
filtered_reason = ['venue.name', 'reasons.items', 'venue.location.lat', 'venue.location.lng']
normalized_reason =normalized_neighborhood.loc[:, filtered_reason]

def get_reason(row):
    try:
        reason_list=row["summary"]
    except:
        reason_list=row["reasons.items"]
    if len(reason_list)==0:
        return None
    else:
        return reason_list[0]["summary"]

# filter the reason for each row
normalized_reason['reasons.items'] = normalized_reason.apply(get_reason, axis=1)

# clean columns
normalized_reason.columns = [col.split(".")[-1] for col in normalized_reason.columns]
```

#### From this review we find that "This spot is popular" is the only content available. We learn that each row is considered as popular, which makes each venue equaliy popular and therefore to be considered in the analysis.
#### Therefore lets continue to split the categories to learn what kind of spots are around.

```python
# filter neighborhoods
filtered_categories = ['venue.id','venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
normalized_neighborhood =normalized_neighborhood.loc[:, filtered_categories]

def get_category_type(row):
    try:
        category_list=row["name"]
    except:
        category_list=row["venue.categories"]
    if len(category_list)==0:
        return None
    else:
        return category_list[0]["name"]

# filter the category for each row
normalized_neighborhood['venue.categories'] = normalized_neighborhood.apply(get_category_type, axis=1)

# clean columns
normalized_neighborhood.columns = [col.split(".")[-1] for col in normalized_neighborhood.columns]
```

#### Now lets plot these categories over the map of Stockholm

```python
nn_r = normalized_neighborhood.loc[normalized_neighborhood['categories'] == 'Restaurant']
nn_b = normalized_neighborhood.loc[normalized_neighborhood['categories'] == 'Bar']
nn_c = normalized_neighborhood.loc[normalized_neighborhood['categories'] == 'Recreation']
nn_s = normalized_neighborhood.loc[normalized_neighborhood['categories'] == 'Shop']
```

```python
# Instantiate a feature group for the restaurants in the dataframe
plotted_stockholm = folium.map.FeatureGroup()

# loop through the venue coordinates for plotting the map
for lat, long, in zip(nn_r.lat, nn_r.lng):
    plotted_stockholm.add_child(
        folium.features.CircleMarker(
            [lat, long],
            radius=7,
            color='black',
            fill=True,
            fill_color='yellow',
            fill_opacity=0.6))

# loop through the bars coordinates for plotting the map
for lat, long, in zip(nn_b.lat, nn_b.lng):
    plotted_stockholm.add_child(
        folium.features.CircleMarker(
            [lat, long],
            radius=7,
            color='black',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6))
    

# loop through the shops coordinates for plotting the map
for lat, long, in zip(nn_s.lat, nn_s.lng):
    plotted_stockholm.add_child(
        folium.features.CircleMarker(
            [lat, long],
            radius=7,
            color='black',
            fill=True,
            fill_color='green',
            fill_opacity=0.6))
    

# loop through the recreation coordinates for plotting the map
for lat, long, in zip(nn_c.lat, nn_c.lng):
    plotted_stockholm.add_child(
        folium.features.CircleMarker(
            [lat, long],
            radius=7,
            color='black',
            fill=True,
            fill_color='red',
            fill_opacity=0.6))
```

#### Average latitude/longitude: There are a few ways to calculate the geographic midpoints, source (http://www.geomidpoint.com/calculation.html) but within a limit of less than 250 miles, averages can be used.

```python
nn_rm = nn_r.groupby('categories').mean()
nn_bm = nn_b.groupby('categories').mean()
nn_cm = nn_c.groupby('categories').mean()
nn_sm = nn_s.groupby('categories').mean()
```
